```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy.integrate
```

```python
data = np.loadtxt("droptower_vdata.txt")
time=data[:,0]
velocity=data[:,1]
position = scipy.integrate.cumtrapz(velocity,time,initial=0)
acceleration = np.diff(velocity)
```

```python
plt.plot(time,position)
plt.plot(time,velocity)
plt.plot(time[:-1],acceleration)
```

```python
pos=acceleration>0
posacc=acceleration[pos]
print(np.mean(posacc))
```
